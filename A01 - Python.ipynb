## Genera una lista de 100 elementos con múltiplos de 3. E.g. [3, 6, 9, ...]

```python

multiplos_de_tres = []
contador = 1

while contador <= 100:
    multiplo = contador * 3
    multiplos_de_tres.append(multiplo)
    contador = contador + 1


filas = 10
elementos_por_fila = 0

for numero in multiplos_de_tres:
    print(numero, end="\t")
    elementos_por_fila = elementos_por_fila + 1

    if elementos_por_fila == filas:
        print("")  
        elementos_por_fila = 0

```

```python

datos_anuales = [
    ("Enero", 1340, 1880),
    ("Febrero", 2590, 3130),
    ("Marzo", 3570, 2140),
    ("Abril", 4450, 3370),
    ("Mayo", 1330, 2990),
    ("Junio", 1610, 1970),
    ("Julio", 1500, 2230),
    ("Agosto", 2270, 1240),
    ("Septiembre", 1910, 1470),
    ("Octubre", 2120, 2280),
    ("Noviembre", 3950, 1870),
    ("Diciembre", 1380, 3140)
]


utilidades = []
for mes, ingreso, gasto in datos_anuales:
    utilidad = ingreso - gasto
    utilidades.append(utilidad)


meses_utilidad_negativa = []
suma_gastos_utilidad_negativa = 0
contador_meses_utilidad_negativa = 0

for i, utilidad in enumerate(utilidades):
    if utilidad < 0:
        meses_utilidad_negativa.append(datos_anuales[i][0])
        suma_gastos_utilidad_negativa += datos_anuales[i][2]
        contador_meses_utilidad_negativa += 1


promedio_gastos_utilidad_negativa = 0
if contador_meses_utilidad_negativa > 0:
    promedio_gastos_utilidad_negativa = suma_gastos_utilidad_negativa / contador_meses_utilidad_negativa


ingresos_totales = sum(ingreso for mes, ingreso, gasto in datos_anuales)


print("Resultados:")
print("-----------")
print("Utilidades por mes:")
for mes, utilidad in zip((mes for mes, _, _ in datos_anuales), utilidades):
    print(f"{mes}: Utilidad = ${utilidad}")
print("-----------")
if contador_meses_utilidad_negativa > 0:
    print(f"Promedio de gastos en meses de utilidad negativa: ${promedio_gastos_utilidad_negativa:.2f}")
else:
    print("No hubo meses con utilidad negativa.")
print(f"Ingresos totales del año: ${ingresos_totales}")
```
