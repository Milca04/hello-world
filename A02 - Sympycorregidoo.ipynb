$f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
import matplotlib.pyplot as plt

sp.var("x")

def f(x):
    return 5 + x + 4/x

df = sp.diff(f(x), x)
pc = sp.solve(df, x)
ddf = sp.diff(df, x)

critical_points_info = []

for c in pc:
    crit_2da_derivada = ddf.subs(x, c)
    
    if crit_2da_derivada < 0:
        point_info = (c, "máximo")
    elif crit_2da_derivada > 0:
        point_info = (c, "mínimo")
    else:
        point_info = (c, "punto de inflexión")
    
    critical_points_info.append(point_info)

for c, info in critical_points_info:
    print("El punto crítico se encuentra en", c, "y es un", info)

xn = np.linspace(0.1, 5, 1000) 

plt.figure(figsize=(6, 4))
plt.plot(xn, f(xn))
for c, _ in critical_points_info:
    plt.plot(c, f(c), "*r", ms=10)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de la función f(x)")
plt.grid(True)
plt.show()


```

El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? ##  1.- Definir las variables simbólicas
##  2.- Restricción de la longitud total
## 3.- Expresión del área de la portería
## 4.- Resolver la restricción para y
## 5.- Sustituir y en la expresión del área
## 6.- Calcular la derivada con respecto a x
## 7.-Resolver da/dx = 0 para encontrar el valor crítico de x
## 8.- Evaluar el valor crítico de x en la expresión de y
## 9.- Calcular el valor del área correspondiente al valor crítico de x
## 10.- Imprimir los resultados
    

```python
import sympy as sp

x, y = sp.symbols('x y')

length_constraint = 2*x + y - 6

area_expression = x * y

y_in_terms_of_x = sp.solve(length_constraint, y)[0]

area_expression_substituted = area_expression.subs(y, y_in_terms_of_x)

da_dx = sp.diff(area_expression_substituted, x)

critical_x = sp.solve(da_dx, x)

critical_y = [y_in_terms_of_x.subs(x, x_val) for x_val in critical_x]

max_area = area_expression_substituted.subs([(x, critical_x[0]), (y, critical_y[0])])

print("Valor crítico de x:", critical_x[0])
print("Valor correspondiente de y:", critical_y[0])
print("Área máxima:", max_area)
```

Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

Definir símbolos para las dimensiones

```python
import sympy as sp
x, y = sp.symbols('x y')
```

Longitud total

```python
longitud_total = 6
```

Restricción de longitud: 2x + y = longitud_total

```python
restriccion_longitud = 2*x + y - longitud_total
```

Expresión para el área de la portería

```python
area_porteria = x * y
```

Sustituir y en la restricción de longitud

```python
y_en_funcion_de_x = sp.solve(restriccion_longitud, y)[0]
```

Sustituir y en la expresión del área

```python
area_en_funcion_de_x = area_porteria.subs(y, y_en_funcion_de_x)
```

Calcular la derivada de área con respecto a x

```python
derivada_area = sp.diff(area_en_funcion_de_x, x)
```

Encontrar el valor crítico de x haciendo derivada = 0

```python
x_critico = sp.solve(derivada_area, x)[0]
```

Encontrar el valor correspondiente de y usando la restricción de longitud

```python
y_critico = y_en_funcion_de_x.subs(x, x_critico)
```

 Calcular el área máxima

```python
area_maxima = area_en_funcion_de_x.subs(x, x_critico)
```

Mostrar los resultados

```python
print("Longitud de los postes (x):", x_critico)
print("Longitud del travesaño (y):", y_critico)
print("Área máxima de la portería:", area_maxima)
```
