1. Resolver numéricamente con `solve_ivp` la siguiente ecuación diferencial:
$$\frac{dx}{dt}=\frac{x}{t+1}$$

```python
import numpy as np 
from scipy.integrate import solve_ivp, odeint
```

```python
def f(t, x) :
    return x/(t+1)
```

Grafique con condición inicial $x_0 = 2$

```python
tspan = [0, 10]
x0 = np.array([2])


x = solve_ivp(f, tspan, x0, max_step =0.1)
x
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.plot(x.t, x.y [0,:], label = "Solución numerica (IVP)")
plt.grid()
plt.legend(loc="best")
```

$$
\frac{d^2 x}{dt^2} +  \omega_{0}^2 x = 0.
$$
Donde $x_0=0.5$, $\dot{x}_0=0.1$ y $\omega_0=3$.

2.1 Defina una función, cuyos argumentos de entrada sean:


1. el tiempo $t$,
2. el arreglo $[x,\dot{x}]$.

Y la salida:

1. el arreglo $[\dot{x},\ddot{x}]$.

  Considerando el argumento de la frecuencia natural $\omega_0$ debe ser una variable global.

```python
w0 = 3
x0 = np.array([0.5, 0.1])
def dx(t, X) :
    x = X[0]
    dx = X[1]
    d2x = -w0**2*x
    return [dx, d2x]

t_span = [0, 20]
j = solve_ivp(dx, t_span, x0, max_step =0.1)
    
```

Grafique

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.plot(j.t, j.y  [0,:], label = "Solución numerica (IVP) Velocidad")
plt.plot(j.t, j.y [1,:], label = "Solución numerica (IVP) Posicion")
plt.grid()
plt.legend(loc="best")
```
